Here we can see the PDF of an exponential distribution whose values are translated by 1. Its mean value is translated by 1 as well.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Set the rate parameter (lambda) for the exponential distribution
lambda_parameter = 1

# Generate values for x
x = np.linspace(0, 5)

# Calculate the PDF for each x using the expon.pdf function
pdf_values = expon.pdf(x-1, scale=1/lambda_parameter)

# Plot the PDF
plt.plot(x, pdf_values, label='Exponential PDF')

# Add labels and a legend
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution PDF')
plt.legend()

# Show the plot
plt.show()
```
